```python
import pandas as pd
```

```python
census_data = pd.read_csv('Census_Data_Cleaned.csv')
census_data.head()
```

```python
burgers_data = pd.read_csv('burgers_data.csv')
burgers_data.head()
```

```python
combined_data = pd.merge(census_data, burgers_data, on="Zip Code")
combined_data.head()
```

```python
pop_desc = combined_data.sort_values(by='Population',ascending=False)
pop_desc.head()
```

```python
mean_rating = combined_data['Burgers Average Rating'].mean()
mean_rating
```

```python
combined_data_filtered = combined_data.loc[(combined_data['Burgers Average Rating'] <= 4)]
combined_data_filtered.head()
```

```python
mean_review = combined_data['Burgers Review Count'].mean()
mean_review
```

```python
combined_data_filtered_review = combined_data_filtered.loc[(combined_data_filtered['Burgers Review Count'] >= 2600)]
combined_data_filtered_review
```

```python
mean_pop = combined_data['Population'].mean()
mean_pop
```

```python
combined_data_filtered_pop = combined_data_filtered_review.loc[(combined_data_filtered_review['Population'] >= 14000)]
combined_data_filtered_pop
```

```python
combined_data_filtered_pop["State"].count()
```

```python
combined_data_filtered_state = combined_data_filtered_pop.loc[(combined_data_filtered_review['State'] == 'CA')]
combined_data_filtered_state
```

```python
import matplotlib.pyplot as plt
import numpy as np

colors = np.array(['red', 'blue', 'green', 'purple'])



graph = combined_data_filtered_state.plot.scatter(x = 'Burgers Review Count',
                                                  y = 'Population',
                                                  c = colors
                                                  )



plt.title('Review Count by Population')
```

```python
import matplotlib.pyplot as plt
import numpy as np

colors = np.array(['red', 'blue', 'green', 'purple'])



graph = combined_data_filtered_state.plot.scatter(x = 'City',
                                                  y = 'Burgers Review Count',
                                                  c = colors
                                                  )



plt.title('Review Count by City')
```
